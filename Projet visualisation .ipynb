```
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```
iris= pd.read_csv('iris.csv')
```

```
iris.head()
```

```
plt.scatter(iris['sepal.length'],iris['sepal.width'])
```

```
sns.pairplot(iris)
```

```
sns.pairplot(iris,hue='variety')
```

```
titanic=sns.load_dataset('titanic')
titanic.head()
```

```
sns.pairplot(titanic)
```

```
sns.catplot(x='pclass', y='age',data=titanic , hue='sex')
```

```
sns.boxplot(x='pclass', y='age',data=titanic , hue='sex')
```

```
sns.displot(titanic['fare'])
```

```
#on peut trouver des correlation entre nos donner 
sns.heatmap(titanic.corr())
```
